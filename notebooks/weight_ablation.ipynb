# Weight Ablation: Sampling vs Loss Weighting

**5 runs:**
1. **Baseline** — uniform weights, no reweighting
2. **Sampler** — `WeightedRandomSampler` with w_min=0.5, w_max=2.0
3. **Loss-weighted** — loss multiplication, w_min=0.5, w_max=2.0
4. **Loss-weighted (zero-floor)** — loss multiplication, w_min=0.0, w_max=2.0
5. **Loss-weighted (negative)** — loss multiplication, w_min=-0.5, w_max=2.0

Config: 50k synthetic, 5k guide, DeBERTa-v3-small, minmax scheme, k=50
